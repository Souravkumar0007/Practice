```python
sns.countplot(x='Gender',data=df)
plt.show()
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.show()
```

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
plt.show()
```

```python
df.columns
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
ax=sns.countplot(data=sales_state,x='State')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
plt.show()
```

```python
sales_state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state_amount,x='State',y='Amount')
plt.show()
```

```python
sales_state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(6)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state_amount,x='State',y='Amount')
ax=sns.countplot(data=sales_state_amount,x='State')
for  bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_marital_amount=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=sales_marital_amount,x='Marital_Status',y='Amount', hue='Gender')
ax=sns.countplot(data=sales_marital_amount,x='Marital_Status')
for bar in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
occupation_data=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=occupation_data,x='Occupation',y='Amount')
plt.show()
```

```python
occupation_data=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=occupation_data,x='Occupation',y='Amount')
ax=sns.countplot(data=occupation_data,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
